```python
def load(filename):
    """
        To use for loading data file to hash tables
        filename: input filename
        HashTable: output of hash tables using dictionary with sets
    """
    
    hashTable = {}
    with open(filename) as file:
        for line in file:
            line = int(line.strip())
            index = hashing(line)
            try:
                hashTable[index].add(line)
            except:
                hashTable[index] = {line}
    return hashTable


def print_hashTable(hashTable):
    for key, value in hashTable.items():
        print(key, end = " ")
        for item in value:
            print("-->", item, end = " ")
        print()


def hashing(number):
    # hash based on absolute value in ten thousands
    return (abs(number) // 10000)


def sum_of_2(hashTable):
    
    """
        Not exactly the 2-sum algorithm, since data set is huge and takes long time to use 2-sum algorithm
        From the hash table take all numbers in buckets: (hash, hash-1, hash+1) into a list.
        Then compute all sums of every two number from this list, and add the sum to a set if it's between -10000 and 10000. 
        
        The logic is that since the hash table is based on absolute value in ten thousand, another amount that makes the sum
        between -10000 and 10000 should be in prior bucket or next bucket.
        For example: 68,037,543,430 is bucketed to key 6803754, to get to the target range, another amt should be between
        (-68,037,553,430, -68,037,533,430) in bucke of key 6803755 and 6803753
    """
    
    result = set()
    
    for key in hashTable.keys():
        lis = []
        lis.extend(hashTable.get(key))
        if hashTable.get(key + 1):
            lis.extend(hashTable.get(key + 1))
        if hashTable.get(key-1):
            lis.extend(hashTable.get(key - 1))
        
        for i in range(len(lis)):
            for j in range(i + 1, len(lis)):
                sums = lis[i] + lis[j]
                if sums >= -10000 and sums <= 10000:
                    result.add(sums)                     
    
    return len(result)



if __name__ == "__main__":
    
    """
    
    The goal of this problem is to implement a variant of the 2-SUM algorithm covered in this week's lectures.
    The file contains 1 million integers, both positive and negative (there might be some repetitions!).
    This is your array of integers, with the i th row of the file specifying the i th entry of the array.
    Your task is to compute the number of target values tt in the interval [-10000,10000] (inclusive) such that 
    there are distinct numbers x,y in the input file that satisfy x+y=t.
    """
    
    import time
    
    start = time.time()
    hashTable = load("2sum.txt")
    print("The count is:", sum_of_2(hashTable))
    end = time.time()
    print(f"The time spent is {end - start} second(s)")


```
